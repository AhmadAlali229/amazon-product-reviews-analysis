```python
import sqlite3

conn = sqlite3.connect("../data/AmazonData.db")
cursor = conn.cursor()

# اطبع الجداول
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in DB:", tables)

# اطبع الأعمدة لجدول AmazonData
cursor.execute("PRAGMA table_info(AmazonData);")
columns = cursor.fetchall()
print("\nColumns in AmazonData:")
for col in columns:
    print("-", col[1])  # col[1] هو اسم العمود

conn.close()
```

```python
##I accedntly Created an Extra table named "employee" from past work, and thats how I fixed it

import sqlite3

conn = sqlite3.connect("../data/AmazonData.db")
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS employees;")

conn.commit()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables left in DB:", tables)

conn.close()
```
